```python
# 11)
numero1 = int(input('Digite o primeiro número inteiro: '))
numero2 = int(input('Digite o segundo número inteiro: '))
numero3 = int(input('Digite o ultimo número inteiro: '))

calculo1 = (numero1 * 2) + (numero2 // 2)
calculo2 = (numero1 * 3) + numero3
calculo3 = numero3 ** 3

print(f'O produto do dobro do primeiro com metade do segundo é {calculo1}n\
, o triplo do primeiro com o terceiro é {calculo2}n\
, o terceiro elevado ao cubo é {calculo3}')
```

```python
altura = float(input('Digite sua altura: '))
peso_ideal = (72.7 * altura) - 58
print(f'Seu peso ideal é {peso_ideal:.1f}')
```
