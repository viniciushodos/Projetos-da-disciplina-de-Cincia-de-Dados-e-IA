```python
#19)
def classifica_nota(nota):
  if nota >= 7:
    print('Aprovado')
  elif nota < 7 and nota >= 5:
    print('Recuperação')
  else:
    print('Reprovado')

classifica_nota(6)
```

```python
#20)
mistos = [-3,-2,-1,0,1,2,3]
positivo = []
negativo = []
zero = []

for i in mistos:
  if i > 0:
    positivo.append(i)
  elif i == 0:
    zero.append(i)
  else:
    negativo.append(i)

  print(f'positivos {positivo}, zero {zero}, e negativo {negativo}')
```
